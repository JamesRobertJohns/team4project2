```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os

sns.set()
```

```python
df = pd.read_csv("fixed_edges_output_1.csv")
df.head()
```

```python
x = np.array(df["NumVertex"])
y = np.array(df["Runtime"])
fig = plt.figure(figsize=(12,10))

plt.plot(x, y)

plt.xlabel('Number of Vertex')
plt.ylabel('Runtime (ms)')
plt.title('Growth of Time against Number of Vertex')
plt.savefig("../docs/pict/Q1_fixedEdges.png")
plt.show()
```

```python
df = pd.read_csv("fixed_vertex_output_1.csv")
df.head()
```

```python
x = np.array(df["NumEdges"])
y = np.array(df["Runtime"])
fig = plt.figure(figsize=(12,10))

plt.plot(x, y)

plt.xlabel('Number of Edges')
plt.ylabel('Runtime (ms)')
plt.title('Growth of Time against Number of Edges')
plt.savefig("../docs/pict/Q1_fixedVertex.png")
plt.show()
```
